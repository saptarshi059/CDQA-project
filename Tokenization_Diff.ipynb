```python
import pickle5 as pickle
from transformers import AutoTokenizer

with open('Metamap_Tokenizations.pkl', 'rb') as f:
    Metamap_Tokenizations = pickle.load(f)
    
bert_tokenizer = AutoTokenizer.from_pretrained('phiyodr/bert-base-finetuned-squad2')
roberta_tokenizer = AutoTokenizer.from_pretrained('roberta-base')
```

```python
bert_tokenizations = []
roberta_tokenizations = []

for row in Metamap_Tokenizations.itertuples(index=False):
    bert_tokenizations.append(bert_tokenizer.tokenize(row.Tokenization, is_split_into_words=True))
    roberta_tokenizations.append(roberta_tokenizer.tokenize(row.Tokenization, is_split_into_words=True))
```

```python
for i in range(len(bert_tokenizations)):
    print(Metamap_Tokenizations.iloc[i].Tokenization)
    print(bert_tokenizations[i])
    print(roberta_tokenizations[i])
    print('...')
    
```

```python
a = set(bert_tokenizer.vocab.keys())
b = set(bert_tokenizer.vocab.keys())
len(b) == len(a) #True
a.intersection(b) == a #True
a.intersection(b) == b #True 
#BERT & RoBERTa's vocab are the same
```

```python
for seq in Metamap_Tokenizations.itertuples(index=False):
    print(seq.Mappings)
    for t in seq.Tokenization:
        print(roberta_tokenizer.tokenize(t))
    print('...')
```
